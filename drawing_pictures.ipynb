```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
names = ['$FinEvent_{no}$', '$FinEvent_{A2C}$', 'FinEvent', '$ReDHAN_{no}$', '$ReDHAN_{AC}$', 'ReDHAN']  # 横坐标 with subscript
index = np.arange(len(names))

NMI_values = ['0.768', '0.784', '0.788','0.751','0.783','0.795']
AMI_values = ['0.680','0.699','0.707','0.659','0.705','0.719']
ARI_values = ['0.324','0.314','0.366','0.713','0.822','0.83']

NMI_values = [float(i) for i in NMI_values]
AMI_values = [float(i) for i in AMI_values]
ARI_values = [float(i) for i in ARI_values]

total_width, n = 0.9, 3  # metric number
width = total_width / n  # 0.3
index = index - width

plt.figure(figsize=(15,8))
plt.bar(index, NMI_values,  width=width, label='NMI')
plt.bar(index + width, AMI_values, width=width, label='AMI', tick_label=names)
plt.bar(index + 2*width, ARI_values, width=width, label='ARI' )

# set 参照线
plt.plot(index, NMI_values, color='red')
# plt.plot(index[-3:], NMI_values[-3:], color='red')

# set axis scale
plt.tick_params(labelsize=15)
# 设置坐标轴刻度
plt.xlim([-0.7, 6.2])
plt.ylim([0.0, 0.9])

# set axis label
font_x = {'family':'Times New Roman',
         'weight': 'normal',
         'size': 20}
plt.ylabel('Values', font_x)
plt.xlabel('Models', font_x)

# set title label
title_x = {'family':'Times New Roman',
         'weight': 'normal',
         'size': 30}
plt.title('The Analysis of Reinforcement Learning', title_x)

# annotate values
for a, b in zip(index, NMI_values):
    plt.text(a, b + 0.002, '%.3f' % b, ha='center', va='bottom', fontsize=10)
for a, b in zip(index + width, AMI_values):
    plt.text(a, b + 0.002, '%.3f' % b, ha='center', va='bottom', fontsize=10)
for a, b in zip(index + 2*width, ARI_values):
    plt.text(a, b + 0.002, '%.3f' % b, ha='center', va='bottom', fontsize=10)

# 显示图例
plt.legend(loc=(0.897,0.009), fontsize=15)  # (0.9,-0.2)
# ['best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center']]
    
# 显示网格
# plt.grid(True,color='k')

# adjust location of graph
# plt.subplots_adjust(left=0.1, bottom=0.1)

plt.show()
```

```python
import numpy as np
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# 假设你有模型的预测概率和真实标签
predicted_probabilities = np.array([0.2, 0.6, 0.7, 0.8, 0.4, 0.9, 0.3])
true_labels = np.array([0, 1, 1, 1, 0, 1, 0])

# 计算ROC曲线数据
fpr, tpr, thresholds = roc_curve(true_labels, predicted_probabilities)
roc_auc = auc(fpr, tpr)

# 绘制ROC曲线
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```
