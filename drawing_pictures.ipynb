```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## fig: para_analy - Accuracy

```python
# 颜色： rainbow, cm.coolwarm, cm.gist_rainbow

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
x_mesh,y_mesh=np.meshgrid(x,y,indexing='ij')#生成网络,参数'ij'生成网格形状是第一个维度对应x，第二个维度对应y
#MCF-7 F-SCORE
z_mesh= np.array([[0.6911, 0.6900,0.6923,0.7011,0.7022],
                 [0.6900,0.6921,0.6956,0.7066,0.7366],
                  [0.6888,0.6911,0.6977,0.7022,0.7152],
                  [0.6826,0.6900,0.6966,0.7123,0.7132],
                  [0.6526,0.6600,0.6566,0.6577,0.6666],
                  ])
```

```python
fig=plt.figure() #这个写一次哦
sub=fig.add_subplot(221,projection='3d')#3d表示三维图像
```

```python
sub.set(xlim=(1, 5), xticklabels=['256','128','64','32','16'],
        ylim=(1, 5), yticklabels=['60','120','180','240','300'],
        zlim=(0.60,0.75), zticks=np.arange(0.60,0.75, 0.05) )
sub.plot_surface(x_mesh,y_mesh,z_mesh,cmap='OrRd',edgecolors='black')
sub.set_xlabel(r'$x$', fontsize=14)
sub.set_ylabel(r'$y$', fontsize=14)
sub.set_zlabel("Accuracy", fontsize=14)
plt.title("Accuracy on PAN2020",fontsize=14)
```
